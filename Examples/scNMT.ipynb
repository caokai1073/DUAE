```python
from unioncom import UnionCom
import numpy as np
import unioncom
from sklearn import preprocessing
```

```python
data1 = np.loadtxt("../scNMT/Paccessibility_300.txt")
data2 = np.loadtxt("../scNMT/Pmethylation_300.txt")
data3 = np.loadtxt("../scNMT/RNA_300.txt")
type1 = np.loadtxt("../scNMT/type1.txt")
type2 = np.loadtxt("../scNMT/type2.txt")
type3 = np.loadtxt("../scNMT/type3.txt")
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
type3 = type3.astype(np.int)
not_connected, connect_element, index = unioncom.utils.Maximum_connected_subgraph(data3, kmax=40)
if not_connected:
	data3 = data3[connect_element[index]]
	type3 = type3[connect_element[index]]
min_max_scaler = preprocessing.MinMaxScaler()
data3 = min_max_scaler.fit_transform(data3)
print(np.shape(data3))
```

```python
uc = UnionCom.UnionCom(epoch_pd=10000, epsilon=0.001, log_pd=1000)
integrated_data = uc.fit_transform(dataset=[data1,data2])
```

```python
uc.test_LabelTA(integrated_data, [type1,type2])
uc.Visualize([data1,data2], integrated_data, mode='PCA') # without datatype
uc.Visualize([data1,data2], integrated_data, [type1,type2], mode='PCA') # with datatype
```

```python
uc = UnionCom.UnionCom(epoch_pd=10000, epsilon=0.001, log_pd=1000)
integrated_data = uc.fit_transform(dataset=[data1,data2,data3])
```

```python
uc.test_LabelTA(integrated_data, [type1,type2,type3])
uc.Visualize([data1,data2,data3], integrated_data, mode='PCA') # without datatype
uc.Visualize([data1,data2,data3], integrated_data, [type1,type2,type3], mode='PCA') # with datatype
```
